```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://oig.justice.gov/reports?keys=&field_publication_date_value=2016-01-01&field_publication_date_value_1=2021-12-20&field_doj_component_target_id=All&field_report_type_target_id=147&field_location_country_code=All&sort_by=field_publication_date_value&sort_order=DESC&items_per_page=250")
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup.text
```

```python
links=[]
for i in soup.findAll('div','views-field views-field-title'):
    final_link = "https://oig.justice.gov" + i.find('a').get('href')
    links.append(final_link)
```

```python
for i in links:
    sub_response=requests.get(i)
    sub_soup=BeautifulSoup(sub_response.content,'html.parser')
    data = sub_soup.findAll('div','mobile-lg:grid-col-9 doj-field-item')
    Date = data[0].text.replace("\n","")
    Report_Number = data[1].text.replace("\n","")
    Component = data[2].text.replace("\n","")
    Pdf_Link = "https://oig.justice.gov" + data[-1].find('a').get('href')
    print(f"[ DATE => {Date} || REPORT NUMBER => {Report_Number} || COMPONENT => {Component} || PDF LINK => {Pdf_Link} ]")
    print("********************\n\n")

    
```
